```python
import numpy as np 
import matplotlib.pyplot as plt
import scipy as sp
import sympy as sy

t_n4_d1= [13.30, 13.00, 13.20, 13.28, 12.89 ]
t_n4_d2= [17.94, 17.37, 18.05, 17.80, 18.07, 17.40]
t_n4_d3= [23.16, 23.08, 23.07, 22.96, 23.07]

t_mean_all=[t_n4_d1, t_n4_d2, t_n4_d3]
t_mean_list=[]
for i in range(len(t_mean_all)):
    t_mean = np.mean(t_mean_all[i])
    t_mean_list.append(t_mean)

print(f" Die Mittelwerte für t : {t_mean_list}")

## Formel für Winkelbeschleunigung:
n = 4
alpha_list = []
flip_alpha_list = []
for i in range(len(t_mean_list)):
    alpha = n * 4 * np.pi / t_mean_list[i]**2 
    alpha_list.append(alpha)
    flip_alpha = 1/alpha
    flip_alpha_list.append(flip_alpha)
    
print(f"Die Winkelbeschleunigungen: {alpha_list}")
print(f"1/alpha Werte sind: {flip_alpha_list}")

d_1 = 23*10**-2
d_2 = 44*10**-2
d_3 = 72*10**-2

d_plot1 = d_1**2
d_plot2 = d_2**2
d_plot3 = d_3**2

d_gesamt_plot = [d_plot1, d_plot2, d_plot3]

##Lineare Regression

linreg=sp.stats.linregress(d_gesamt_plot, flip_alpha_list)
print(linreg)
m=linreg.slope
t=linreg.intercept
x=np.linspace(min(d_gesamt_plot), max(d_gesamt_plot), 501)
y=m*x+t


reglabel=f' $y(x) = m \\cdot x + t$ \n $m$={m:.6f} \n $t$={t:.6f}'
plt.plot(x, y, label=reglabel)
plt.plot(d_gesamt_plot, flip_alpha_list , linestyle='', marker='+', label="Messwerte")
plt.title('1/alpha zu d^2')
plt.legend()
plt.xlabel("d^2")
plt.ylabel('1/alpha')
plt.grid(visible=True, color='black', linewidth=0.5)


plt.show()
##for i in range t

##alpha = 
##M = m_zug* g * r_1
##alpha * J = M
##alpha = M/J
##1/a = J/M
##J = (1/12 * m_s * L^2) + (1/2 * m_H * d^2)


```
